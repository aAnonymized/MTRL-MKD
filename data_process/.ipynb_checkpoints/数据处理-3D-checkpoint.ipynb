## initialize an ROI (Region of Interest) extraction network

```python
unet = UNet(3, 4)  # channel=3, seg_num=4
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
unet = unet.to(device)

pretrained_dict = torch.load('./unet_model/epoch_325.pth', map_location=device)
model_dict = unet.state_dict()
# 过滤掉不匹配的参数
pretrained_dict = {k: v for k, v in pretrained_dict.items() if k in model_dict and v.size() == model_dict[k].size()}
model_dict.update(pretrained_dict)
unet.load_state_dict(pretrained_dict, strict=False)
```

### Contrust ACDC dataset

```python
# origin dataset dir.
train_pp = r'./database/training'
test_pp = r'./database/testing/'

## save dataset dir.
save_train_pp = r'./processed/training'
save_test_pp = r'./processed/testing'
os.makedirs(save_test_pp, exist_ok=True)
os.makedirs(save_train_pp, exist_ok=True)
```

```python
#### create train dataset
patient_data = {}
for patient_dir in os.listdir(train_pp):
    with open(os.path.join(train_pp, patient_dir, 'Info.cfg')) as f_in:
        for line in f_in:
          l = line.rstrip().split(": ")
          patient_data[l[0]] = l[1]
    label = patient_data['Group']
    
    source_path_list = []
    source_path_list.append(os.path.join(train_pp, patient_dir, f'{patient_dir}_4d.nii.gz'))

    target_path = os.path.join(save_train_pp, label)
    os.makedirs(target_path, exist_ok=True) # 不存在创建.
    for source_path in source_path_list:
        shutil.copy(source_path, target_path)
```

### 主函数

```python
unet.eval()

for index, pp in enumerate(pppp):
    label_dirs = os.listdir(pp)
    save_pp = save_pppp[index]
    os.makedirs(save_pp, exist_ok=True) # 不存在创建.
    # with open(os.path.join(save_pp, f'{csv_list[index]}'), mode="a", newline="", encoding="utf-8") as outfile:
    #     writer = csv.writer(outfile)                         
    #     writer.writerow(['full_path', 'Finding Labels'])
    for label_dir in label_dirs:
        os.makedirs(os.path.join(save_pp, label_dir), exist_ok=True) # 不存在创建.
        for file_index, file_name in enumerate(os.listdir(os.path.join(pp, label_dir))):
            # try:
            print(f"Now processing {file_name} .....")
            image = nib.load(f'{pp}\{label_dir}\{file_name}')
            origin_image_array = image.get_fdata()  ## W H S T
            hdr = image.header
            spacing = hdr.get_zooms()
            spacing = spacing[:3]
            print(f'spacing : {spacing}')
            origin_image_array=np.transpose(origin_image_array, (-1, 2, 1, 0))  ## T S W H 
            
            ### select 3 slice.
            T, S, W, H = origin_image_array.shape
            m = S //2
            image_array = origin_image_array[:, [m-2, m, m+2], :, :]  ## T 3 W H
            
            # unify spacing.
            resampled_frames = []
            new_spacing = None
            for t in range(image_array.shape[0]):
                resampled_frame, spacing = standardize_resolution(image_array[t], spacing)
                resampled_frames.append(resampled_frame)
            image_array = np.stack(resampled_frames)  ## T 3, W H
            T, S, W, H = image_array.shape
            new_spacing = spacing

            ### extractor ROI
            roi_image_list = []
            for S in range(image_array.shape[1]):
                image_s = image_array[0, S, :, :]  # T=0表示ED, ED 的效果是最好.
                pred_map = predict_full_image_argmax(model=unet, image=image_s, cmap=None, device='cpu')  ### 建议改写成能同一帧, 不同 slice 组成 B 3 w h 的批次同时推理.
                center_x, center_y = get_LV_center(pred_map)
                if center_x is not None and center_y is not None:
                    x1, y1, x2, y2 = get_bbox(center_x, center_y, W, H, crop_w, crop_h)
                    roi_image_list.append(image_array[:, S, y1:y2, x1:x2])
                    if 0:
                        fig, ax = plt.subplots()
                        ax.imshow(image_array[0, S, :, :][0], cmap='gray')
                        rect = patches.Rectangle((x1, y1), x2 - x1, y2 - y1,
                                                 linewidth=2, edgecolor='red', facecolor='none')
                        ax.add_patch(rect)
                        plt.title("Prediction Box")
                        plt.axis('off')
                        plt.show()
            
            roi_image_array = np.stack(roi_image_list, axis=1)

            # imnormalize
            roi_image_array = z_score_normalize(roi_image_array)
            if 0:
                print(f"roi_image_array : {roi_image_array.shape}")
                fig, ax = plt.subplots()
                ax.imshow(roi_image_array[roi_image_array.shape[0]//2, roi_image_array.shape[1]//2, :, :], cmap='gray')
                plt.title("Prediction Box")
                plt.axis('off')
                plt.show()
            
            # output = f'{save_pp}/{label_dir}/{S}_{file_name}'
            # image = sitk.GetImageFromArray(roi_image_array)
            # sitk.WriteImage(image, output) # 写入
            # with open(os.path.join(save_pp, f'{csv_list[index]}'), mode="a", newline="", encoding="utf-8") as outfile:
            #     writer = csv.writer(outfile)                         
            #     writer.writerow([output, label_dir])
            # # # except Exception as e:
            # #     pass
```
